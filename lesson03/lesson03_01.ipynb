```python
#class Person:    也可以這樣寫
# def __init__(self,name:str,age:int): #可寫可不寫 type hint
#        self.name = name
#        self.age = age

class Person(object):
    def __init__(self,name,age):
        self.name = name
        self.age = age
    #自訂實體被print()時的輸出
    def __repr__(self) -> str:      #預設傳出值是字串
        return f'My name is {self.name}\nMy age is {self.age}'
    
```

```python
#初始化(initial)
p1 = Person("Ken",20)
print(type(p1))
print(p1.name)
print(p1.age)
print(p1)

print("======================")
p2 =Person("Ben",25)
print(type(p2))
print(p2.name)
print(p2.age,type(p2.age))
print(p2)
```

```python
print(p1.name,p1.age)
print(type(p1.name),type(p1.age))
print('==========================')
print(p2.name,p2.age)

```

```python
#正常情形 不可有重複名稱的 clss
#剛好 jupyter可以 然後 後執行 的會取代 前面的

#通常 __name 有加 __ 會視為私人的 
#搭配 property 就不會被修改
# 所有在class 裡面的 def 都一定要有一個 self 的參數

class Person(object):
    def __init__(self,name,age):
        self.__name = name
        self.__age = age
    #自訂實體被print()時的輸出
    def __repr__(self) -> str:      #預設傳出值是字串
        return f'My name is {self.name}\nMy age is {self.age}'
    
    @property
    def name(self):
        return self.__name
    def age(self):
        return self.__age
    
    #如果下面這行執行時，當有人想要修改 就會類似跳出通知
    @name.setter
    def name(self,n):
        print(f'Not allow to change name to {n}')
```
